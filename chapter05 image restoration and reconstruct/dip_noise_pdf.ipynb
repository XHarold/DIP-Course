# 噪声及其分布

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 高斯分布
$$
f(x)=\frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

```python
mu = 0.5
sigma = 1
N = 10000
x = np.random.normal(mu,sigma,N)
n,bins,_= plt.hist(x,bins=256,density=True)
y = 1/(np.sqrt(2*np.pi)*sigma)*np.exp(-(bins-mu)**2/(2*sigma**2))
plt.plot(bins,y,'r')
```

## 指数分布
$$
p(z)=\left \{
             \begin{array}{lr}
             ae^{-az} & z>0  \\
             0 & z<0
             \end{array}
\right.
$$

```python
alpha = 1
x = np.random.exponential(alpha,N)
n,bins,_= plt.hist(x,bins=256,density=True)
y = alpha*np.exp(-alpha * bins)
plt.plot(bins,y,'r')
```

## Gamm分布
$$
p(z)=\left \{
             \begin{array}{lr}
             \frac{a^bz^{b-1}}{(b-1)!}e^{-az} & z>0  \\
             0 & z<0
             \end{array}
\right.
$$

```python
import scipy.special as sps
```

```python
shape = 2
scale = 3
x = np.random.gamma(shape,scale,N)
n,z,_= plt.hist(x,bins=256,density=True)
y = np.power(z,shape-1) * (np.exp(-z/scale) / (sps.gamma(shape)*scale ** shape))
plt.plot(z,y,'r')
```
