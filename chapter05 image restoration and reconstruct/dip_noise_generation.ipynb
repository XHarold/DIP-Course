# 生成指定分布的噪声

本示例利用skimage内置的函数生成不同类型的噪声

```python
import matplotlib.pyplot as plt
from skimage import data, img_as_float,io
from skimage.util import random_noise
from skimage.color import rgb2gray

%matplotlib inline
```

```python
#对图像添加五种不同的噪声
original = rgb2gray(data.chelsea())
noisy1 = random_noise(original,'gaussian')
noisy2 = random_noise(original,'s&p')
noisy3 = random_noise(original,'localvar')
noisy4 = random_noise(original,'poisson')
noisy5 = random_noise(original,'speckle')


plt.figure()
plt.subplot(231)
plt.imshow(original,cmap="gray")
plt.title('original')
plt.subplot(232)
plt.imshow(noisy1,cmap="gray")
plt.title('gaussian')
plt.subplot(233)
plt.imshow(noisy2,cmap="gray")
plt.title('s&p')
plt.subplot(234)
plt.imshow(noisy3,cmap="gray")
plt.title('localvar')
plt.subplot(235)
plt.imshow(noisy4,cmap="gray")
plt.title('poisson')
plt.subplot(236)
plt.imshow(noisy5,cmap="gray")
plt.title('speckle')
```
