# Winer滤波

本示例展示用Winer滤波去除图像模糊

```python
import numpy as np
import random
from matplotlib import pyplot as plt
from skimage import color, data, restoration,io
import math
import scipy.signal
```

给图像添加高斯噪声
$$
f(x)=\frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

```python
def gaussian_white_noise(img, mu, sigma):
    '''
    加性高斯白噪声（适用于灰度图）
    :param img: 输入图像
    :param mu: 均值
    :param sigma: 标准差
    :return:
    '''
    img_cp = np.copy(img)  # 输入图像的副本

    img_cp += np.random.normal(mu,sigma,img.shape)

    return img_cp

def inverse_filtering(img, h):
    '''
    逆滤波
    :param img: 输入信号
    :param h: 退化函数（时域）
    :return: 滤波后的信号（幅值）
    '''
    output_img = []     #输出图像
    output_img_fft = [] #输出图像的傅里叶变换

    h_fft = np.fft.fft2(h) # 退化函数的傅里叶变换

    img_fft = np.fft.fft2(img)  # 输入图像的傅里叶变换

    output_img_fft = img_fft / h_fft

    output_img = np.abs(np.fft.ifft2(output_img_fft))

    return output_img


def wiener_filtering(img, h, K):
    '''
    维纳滤波
    :param img: 输入信号
    :param h: 退化函数（时域）
    :param K: 参数K
    :return: 维纳滤波后的信号（幅值）
    '''
    output_img = []  # 输出信号

    output_img_fft = [] # 输出信号的傅里叶变换

    img_cp = np.copy(img) # 输入信号的副本

    img_cp_fft = np.fft.fft2(img_cp)  # 输入信号的傅里叶变换

    h_fft = np.fft.fft2(h) # 退化函数的傅里叶变换

    h_abs_square = np.abs(h_fft)**2 # 退化函数模值的平方

    # 维纳滤波
    output_img_fft = np.conj(h_fft) / (h_abs_square + K)

    output_img = np.abs(np.fft.ifft2(output_img_fft * img_cp_fft)) # 输出信号傅里叶反变换

    return output_img
```

```python
img = color.rgb2gray(data.camera())

# 滤波器
h_org = np.eye(20) / 5

# 图像卷积
img_blurring = scipy.signal.convolve2d(img, h_org, mode = 'same')

# 增加加性高斯白噪声
img_with_noise = gaussian_white_noise(img_blurring, 0, 10)

# 求解退化函数h
img_fft = np.fft.fft2(img)
img_with_noise_fft = np.fft.fft2(img_with_noise)
h_pred = np.fft.ifft2(img_with_noise_fft / img_fft)

# 逆滤波
inverse_filtering = inverse_filtering(img_with_noise, h_pred)

# 维纳滤波
K = 0.1
wiener_filtering = wiener_filtering(img_with_noise, h_pred, K)

plt.figure()
plt.subplot(2, 2, 1)
plt.title("Source Image")
plt.imshow(img, cmap="gray")

plt.subplot(2, 2, 2)
plt.title("Degraded Image")
plt.imshow(img_with_noise, cmap="gray")

plt.subplot(2, 2, 3)
plt.title("Inverse filtering")
plt.imshow(inverse_filtering, cmap="gray")

plt.subplot(2, 2, 4)
plt.title("Wiener filtering")
plt.imshow(wiener_filtering, cmap="gray")

plt.subplots_adjust(wspace=0.2,hspace=0.4)
plt.show()
```
