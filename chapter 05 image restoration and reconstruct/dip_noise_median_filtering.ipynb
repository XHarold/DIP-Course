# 利用中值滤波器滤除噪声

```python
import matplotlib.pyplot as plt
from skimage import data, img_as_float,io,filters
from skimage.util import random_noise
from skimage.color import rgb2gray
from skimage.morphology import disk
```

```python
original  = data.camera()
noisy = random_noise(original,'s&p')
edges1= filters.median(noisy,disk(3))
edges2= filters.median(noisy,disk(5))

plt.figure()
plt.subplot(221)
plt.imshow(original,cmap="gray")
plt.title('original')
plt.axis('off')
plt.subplot(222)
plt.imshow(noisy,cmap="gray")
plt.title('s&p noisy')
plt.axis('off')
plt.subplot(223)
plt.imshow(edges1,cmap="gray")
plt.title('median-filter-3')
plt.axis('off')
plt.subplot(224)
plt.imshow(edges2,cmap="gray")
plt.title('median-filter-5')
plt.axis('off')

plt.subplots_adjust(hspace=0.2)
```
