# 图像卷积

卷积是图像处理的基本操作，卷积的通用计算公式是：

$$g(x,y)=\sum_{m=-M}^{M}\sum_{n=-N}^{N}f(x+m,y+n)w(m,n)$$

$g(x,y)$是卷积后的图像，$f(x,y)$是原始图像，$W$是一个窗函数。

卷积本质上是将每个像素点的值替换为它与其相邻元素点的值的加权和。

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.ndimage.interpolation import zoom  
from scipy.signal import convolve2d    # 引入二维卷积函数

im = plt.imread("images/lenna.jpg") # 加载当前文件夹中的图片
print(im.shape) # 输出图像尺寸
```

```python
def convolution(im, window):
    """
    用窗口window卷积图像，依次对图像的每个通道卷积
    """
    
    # 判定图像是灰度图像还是彩色图像
    if len(im.shape) == 2:
        ch = 1
        filtered_im = convolve2d(im, window, mode="same", boundary="symm")
    else:
        ch = 3
        
        filtered_im = []
        for d in range(3):
            # 对图像的三个通道循环处理
            temp = convolve2d(im[:,:,d], window, mode="same", boundary="symm")
            # mode决定输出尺寸
            #boundary决定边界条件
            filtered_im.append(temp)

        filtered_im = np.stack(filtered_im, axis=2).astype("uint8")
    return filtered_im
```

```python
def pltimage(im, titlename, **kwargs):
    """
    画图的辅助函数
    """
    plt.imshow(im, interpolation="none", **kwargs)
    plt.axis('off') # 去掉坐标轴
    plt.title(titlename)
    plt.show()      # 弹窗显示图像
```

```python
im_small = zoom(im, (0.5,0.5,1))
# zoom 将图片每一维以相应系数缩小
# im.shape = (3400, 2000, 3)
# im_small.shape = (680, 400, 3)

n=5
window = np.ones((n,n))
# 构建nxn的全1矩阵
window /= np.sum(window)
# 使矩阵所有元素的和为1
pltimage(convolution(im_small, window),'Filtered')
pltimage(im,'Original')
```
