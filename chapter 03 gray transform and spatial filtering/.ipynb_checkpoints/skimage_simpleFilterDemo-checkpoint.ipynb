```python
from skimage import data,filters
import matplotlib.pyplot as plt
%matplotlib inline

img = data.camera()
edges = filters.sobel(img)
plt.imshow(edges,plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
from skimage import data,filters,feature
import matplotlib.pyplot as plt
%matplotlib inline

img = data.camera()
edges1 = feature.canny(img)   #sigma=1
edges2 = feature.canny(img,sigma=3)   #sigma=3

plt.figure('canny',figsize=(8,8))
plt.subplot(121)
plt.title('sigma=1')
plt.axis('off')
plt.imshow(edges1,plt.cm.gray)  

plt.subplot(122)
plt.title('sigma=3')
plt.axis('off')
plt.imshow(edges2,plt.cm.gray)

plt.show()
```

```python
from skimage import data,filters
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

img = data.camera()
filt_real, filt_imag = filters.gabor(img,frequency=0.6)   
plt.figure('gabor',figsize=(8,8))

plt.subplot(121)
plt.title('Original')
plt.imshow(img,plt.cm.gray)  

plt.subplot(122)
plt.title('filt-imag')
filt_img = filt_real**2+filt_imag**2

plt.imshow(filt_img,plt.cm.gray)

plt.show()
```

```python
from skimage import data,filters
import matplotlib.pyplot as plt
%matplotlib inline

img =data.camera()
dst_neg =filters.roberts_neg_diag(img) 
dst_pos =filters.roberts_pos_diag(img) 

plt.figure('filters',figsize=(8,8))
plt.subplot(131)
plt.title('origin image')
plt.imshow(img,plt.cm.gray)
plt.axis('off')

plt.subplot(132)
plt.title('neg-filted image')
plt.imshow(dst_neg,plt.cm.gray)
plt.axis('off')

plt.subplot(133)
plt.title('pos-filted image')
plt.imshow(dst_pos,plt.cm.gray)
plt.axis('off')
plt.show()
```
