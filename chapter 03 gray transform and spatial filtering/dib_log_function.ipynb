# $log^x$ 与 $y=x$

本示例给了$y=log^x,y=x,y=\sqrt{x}$函数曲线的差异

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,100)
```

```python
y = np.log(x+1e-10)
a = np.min(y)
b = np.max(y)
y = (y - a)/(b - a)
```

```python
%matplotlib inline
```

```python
plt.plot(x,y,linestyle='-', c='red', label='$y=log^{x}$')
plt.plot(x,x,linestyle='--',c='blue',label='$y=x$')
plt.plot(x,np.sqrt(x),linestyle='--',c='green',label='$y=\sqrt{x}$')
plt.legend()
```
