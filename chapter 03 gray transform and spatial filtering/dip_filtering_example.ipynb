# 图像空间滤波

 skimage的filters模块提供了常见的滤波函数，能用于图像的空间滤波。

```python
from skimage import data,color,filters,feature,img_as_float
import matplotlib.pyplot as plt
#inline matploglib
```

```python
img = data.camera()
plt.subplot(121)
plt.imshow(img,cmap=plt.cm.gray)
plt.axis('off')
plt.title('Original')
out = filters.gaussian(img,sigma=3)
plt.subplot(122)
plt.imshow(out,cmap=plt.cm.gray)
plt.axis('off')
plt.title('Smooth')
```

```python
img = data.camera()
plt.subplot(131)
plt.imshow(img,cmap=plt.cm.gray)
plt.axis('off')
plt.title('Original')
out1 = feature.canny(img_as_float(img),sigma=1) 
plt.subplot(132)
plt.imshow(out1,cmap=plt.cm.gray)
plt.axis('off')
plt.title('Canny Sigma=1')
out3 = feature.canny(img_as_float(img),sigma=3) 
plt.subplot(133)
plt.imshow(out3,cmap=plt.cm.gray)
plt.axis('off')
plt.title('Canny Sigma=3')
```
