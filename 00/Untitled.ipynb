```python
img = io.imread('images/640-9.jpeg')
plt.imshow(img,plt.cm.gray)
plt.show()
```

```python
img.shape
```

```python
red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]
```

```python
plt.subplot(2,2,1)
plt.imshow(img)
plt.title('original')

plt.subplot(2,2,2)
plt.imshow(red,cmap=plt.cm.gray)
plt.title('red')

plt.subplot(2,2,3)
plt.imshow(green,cmap=plt.cm.gray)
plt.title('green')

plt.subplot(2,2,4)
plt.imshow(blue,cmap=plt.cm.gray)
plt.title('blue')
```
