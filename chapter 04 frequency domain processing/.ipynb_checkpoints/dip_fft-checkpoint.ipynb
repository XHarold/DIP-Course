```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
%matplotlib inline
```

```python
img = io.imread('images/a.tif')
fft2 = np.fft.fft2(img)
shift_fft2 = np.fft.fftshift(fft2)
```

```python
plt.imshow(img,'gray')
plt.title('original')
plt.axis('off')
```

```python
plt.imshow(np.abs(fft2),'gray')
plt.title('fft2')
plt.axis('off')
```

```python
plt.imshow(np.abs(shift_fft2),'gray')
plt.title('shifted fft')
plt.axis('off')
```

```python
log_shift_fft2 = np.log(1 + np.abs(shift_fft2))
plt.imshow(log_shift_fft2,'gray')
plt.title('log_shift_fft2')
plt.axis('off')
```
