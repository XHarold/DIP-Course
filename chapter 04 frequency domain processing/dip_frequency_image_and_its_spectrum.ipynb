# 图像DFT变换

本示例展示矩形框图像的DFT变换结果，目的是体现空间变化与频域的关系：空间变化越快，频谱越宽；空间变化越慢，频谱越窄

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import scipy.signal as signal

%matplotlib inline
```

```python
N = 256
h_img = np.zeros((N,N),dtype='float')
v_img = np.zeros((N,N),dtype='float')
h_img[112:144,64:192] = 255
v_img[64:192,112:144] = 255
h_fft = np.fft.fftshift(np.fft.fft2(h_img))
v_fft = np.fft.fftshift(np.fft.fft2(v_img))

plt.subplot(221)
plt.imshow(h_img,plt.cm.gray)
plt.title('horizontal bar')
plt.axis('off')

plt.subplot(222)
plt.imshow(v_img,plt.cm.gray)
plt.title('Vertical bar')
plt.axis('off')

plt.subplot(223)
plt.imshow(np.log10(1+abs(h_fft)),plt.cm.gray)
plt.title('fft of horizontal bar')
plt.axis('off')

plt.subplot(224)
plt.imshow(np.log10(1+abs(v_fft)),plt.cm.gray)
plt.title('fft of vertical bar')
plt.axis('off')
```
