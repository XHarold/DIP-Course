# 信号长度对FFT变换的影响

本示例展示了取样点数对信号FFT计算的影响。
$$
F(u)=\mathcal{F}\{f(t)\}=\int_{-\infty}^{\infty}f(t)e^{-j2\pi ut}dt
$$
信号截断，相当于$f(t)u(t)$，对应的频谱为$F(u)\otimes U(u)$,$U(u)$就是一个$sinc$函数，其宽度与$u(t)$宽度成反比，宽度越长，函数约窄。

结论：时间长度影响频域分辨率

```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
fs = 100
duration = 0.5
t = np.linspace(0,duration,fs * duration)
s = 3 * np.cos(2*np.pi*5*t)+2*np.cos(2*np.pi*10*t)+np.cos(2*np.pi*15*t)
plt.plot(t,s)
```

```python
u = np.linspace(-fs/2,fs/2,duration*fs)
f =  np.fft.fftshift(abs(np.fft.fft(s)))
plt.plot(u,f)
```
