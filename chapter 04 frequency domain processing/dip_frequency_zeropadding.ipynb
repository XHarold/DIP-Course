# 信号补零对频谱的影响

```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
M = 256
N = 512
x1 = np.arange(M)
x2 = np.arange(N)
s1 = np.array([random.gauss(0,0.5) for i in range(M)])
s2 = np.zeros((N,))
s2[:M] = s1[:M]

f1 = np.fft.fftshift(abs(np.fft.fft(s1)))
f2 = np.fft.fftshift(abs(np.fft.fft(s2)))

plt.subplot(121)
plt.plot(x1,s1)
plt.title('s1')

plt.subplot(122)
plt.plot(x2,s2)
plt.title('s2')
```

```python
fe = np.zeros((N,))
for i in range(M):
    fe[2*i] = f1[i]

plt.figure(figsize=(8,6))


plt.subplot(211)
plt.stem(x2[0:32], fe[0:32], markerfmt="C2.")
plt.subplot(212)
plt.stem(x2[0:32], f2[0:32], markerfmt="C3.")
```
