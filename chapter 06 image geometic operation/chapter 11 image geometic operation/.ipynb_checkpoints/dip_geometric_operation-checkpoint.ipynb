# 图像几何变换
   图像几何变换是指将位于$(x,y)$位置的像素移动到$(x',y')$位置，像素值不变。图像像素位置的空间关系的变化可以用函数来表示，即：
   $$
   x'=T_x(x,y),y'=T_y(x,y)
   $$
   根据变换函数是否为线性函数可以分为线性变换或非线性变换两种，典型的线性变换为：仿射变换或投影变换，变换前后的位置关系可以表示为:
   $$
   \begin{bmatrix}x'\\y'\\1 \end{bmatrix}=\begin{bmatrix}a_{11} & a_{12} & t_x \\ a_{21}&a_{22}&t_y\\ 0 & 0 & 1 \\ \end{bmatrix}\begin{bmatrix}x\\y\\1\\\end{bmatrix}
   $$
   非线性变换说明变换前后像素的空间位置不能用线性函数的表示。
   
   Skimage中transform模块用于实现图像的线性几何变换。基本方法是：
   
   * 首先根据平移、旋转和缩放构建变换矩阵
   * 其次是利用warp函数实现变换
   
   由于变换会导致图像尺寸的变化，下面这个函数能根据变换矩阵计算变换后图像四个角的位置，进而计算得到图像变换后的高度、宽度和左上角偏离$(0,0)$的距离。
   
  
   这种方法看上去比较简单，但在实际实现时存在一个问题就是$(x,y)$位置的像素经过变换后位置变为$(x',y')$,它的值很可能不是整数，这个时候就需要插值运算，插值运算需要知道$(x',y')$周围整像素位置的值，而通过正变换很可能无法直接获得整像素位置的值。鉴于上述情况，在图像几何变换时通常都是采用逆变换函数形式，即：根据$(x',y')$位置反算其对应的$(x,y)$,比如:根据$(x'=102,y'=254)$计算对应的$(x,y)$，显然$(x,y)$很可能不是整数，也需要插值，由于原始图像整像素位置的值是已知的，因此插值算法是能够实现。
     
   采用逆函数给计算带来了方便，同时也带来了另外一个问题：那就是$x',y'$的取值范围如何来确定？因为我们在计算时需要设定循环的起点与终点，比如： for x' in np.arange(start,end)。通常情况下变换会导致图像尺寸的变化，也就是说输出图像的大小很可能与原始图像不同，并且由于平移等因素，图像的左上角未必就是（0，0）。为了准确获知变换输出图像的位置的大小，我们设计了下面这个函数。它将用(0,0)(w,0)(0,h)(w,h)表示原始图像的外接矩形的四个顶点，通过正变换得到变换后的四个角的位置，进而计算得到图像变换后的高度、宽度和左上角偏离$(0,0)$的距离。利用左上角的偏移我们可以在设计变换矩阵时对其进行反向平移，是的变换后的图像的像素位置$x',y'$落正值区域，同时也可以设定输出图像的宽度和高度。
   

```python
from skimage import io, data, color, transform
import matplotlib.pylab as plt
import numpy as np
#inline matplotlib

def trans_and_box(matrix, shape):
    h = shape[0]
    w = shape[1]
    rect_original = np.array([[0,0,1],[0,h,1],[w,0,1],[w,h,1]])  #构建四个顶点齐次坐标
    rect_mapped = np.dot(matrix, rect_original.T)  #利用矩阵运算计算变换后的顶点位置
    min_val = rect_mapped.min(axis=1)  #计算两个方向的最小值
    max_val = rect_mapped.max(axis=1)  #计算两个方向的最大值
    #计算反向平移的值
    trans_pos = (-np.floor(min_val[0]) if (min_val[0]>0) else -np.ceil(min_val[0]), 
                 -np.floor(min_val[1]) if (min_val[1]>0) else -np.ceil(min_val[1]))
    #计算图像变换后高度和宽度，注意skimage中图像采用(h,w,c)方式来表示
    output_shape = (int(np.ceil(max_val[1]-min_val[1])), \
                    int(np.ceil(max_val[0]-min_val[0])))
    return trans_pos, output_shape
```

## 相似变换(保角)变换
   变换有平移和旋转和缩放组成，其中缩放要求XY两个方向为相同比例

```python
img = data.camera()

tform = transform.SimilarityTransform(scale=2, rotation=np.pi / 4,
                               translation=(0, 0))

trans_pos, output_shape = trans_and_box(tform.params, img.shape)

tform = tform + transform.SimilarityTransform(translation=trans_pos)

img_forward  = transform.warp(img, tform.inverse,output_shape=output_shape)
img_backward = transform.warp(img_forward, tform,output_shape=img.shape)

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(8, 3))

ax1.imshow(img)
ax1.set_title('Original')
ax1.axis('off')
ax2.imshow(img_forward)
ax2.set_title('Similarity')
ax2.axis('off')
ax3.imshow(img_backward)
ax3.set_title('Inverse')
ax3.axis('off')
```

## 仿射变换
   与相似变换唯一不同在于XY方向的放大倍数可以不同

```python
img = data.camera()

tform = transform.AffineTransform(scale=(0.6,1.2), rotation=np.pi/6, shear=np.pi/8)

trans_pos, output_shape = trans_and_box(tform.params, img.shape)

tform = tform + transform.SimilarityTransform(translation=trans_pos)

img_affine  = transform.warp(img, tform.inverse,output_shape=output_shape)
img_inverse = transform.warp(img_affine, tform,output_shape=img.shape)

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(8, 3))

ax1.imshow(img)
ax1.set_title('Original')
ax1.axis('off')
ax2.imshow(img_affine)
ax2.set_title('Affine')
ax2.axis('off')
ax3.imshow(img_inverse)
ax3.set_title('Inverse')
ax3.axis('off')
```

## 投影变换
   投影变换会导致平行线相交

```python
img = data.camera()

# slope of the perspective
slope = np.tan(np.radians(23))
h, w = img.shape
matrix = np.array([[1, 0, 0],
                   [-slope/2, 1, slope * h / 2],
                   [-slope/w, 0, 1 + slope]])

tform = transform.ProjectiveTransform(matrix = matrix)
trans_pos, output_shape = trans_and_box(tform.params, img.shape)

tform = tform + transform.SimilarityTransform(translation=trans_pos)

img_project  = transform.warp(img, tform.inverse,output_shape=output_shape)
img_inverse = transform.warp(img_project, tform,output_shape=img.shape)

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(8, 3))

ax1.imshow(img)
ax1.set_title('Original')
ax1.axis('off')
ax2.imshow(img_project)
ax2.set_title('Project')
ax2.axis('off')
ax3.imshow(img_inverse)
ax3.set_title('Inverse')
ax3.axis('off')
```

## 非线性变换的示例1

   这个例子展示了利用非线性变换实现波浪效果。
   $$
   \begin{align}
   x &=x‘ \\
   y &=y’ + 20sin(2\pi x‘/180) \\
   \end{align}
   $$
   这里20表示幅度，180表示每180点为一个周期
   
   注意：这里同样采用了逆函数的表示方式
      $$
   x=T_x^{-1}(x',y'),y=T_y^{-1}(x',y')
   $$
   
  

```python
from skimage import io, data, color
import matplotlib.pylab as plt
import numpy as np
#inline matplotlib

coffee = data.coffee()
img = color.rgb2gray(coffee)
rows, cols = img.shape

io.imshow(img)
plt.subplot(221)
plt.imshow(img,cmap=plt.cm.gray)
plt.title('Original')
plt.axis('off')

#####################
# Vertical wave

img_vertical = np.zeros(img.shape, dtype=img.dtype)

for i in range(rows):
    for j in range(cols):
        offset_y = int(25.0 * np.sin(2 * 3.14 * i / 180))
        if j+offset_y < cols:
            img_vertical[i,j] = img[i,(j+offset_y)%cols]
        else:
            img_vertical[i,j] = 0
            
plt.subplot(222)
plt.imshow(img_vertical,cmap=plt.cm.gray)
plt.title('Vertical')
plt.axis('off')
#####################
# Horizontal wave

img_horizontal = np.zeros(img.shape, dtype=img.dtype)

for i in range(rows):
    for j in range(cols):
        offset_x = 0
        offset_y = int(16.0 * np.sin(2 * 3.14 * j / 150))
        if i+offset_y < rows:
            img_horizontal[i,j] = img[(i+offset_y)%rows,j]
        else:
            img_horizontal[i,j] = 0
            
plt.subplot(223)
plt.imshow(img_horizontal, cmap=plt.cm.gray)
plt.title('horizontal')
plt.axis('off')

#####################
# Both horizontal and vertical 

img_both = np.zeros(img.shape, dtype=img.dtype)

for i in range(rows):
    for j in range(cols):
        offset_x = int(20.0 * np.sin(2 * 3.14 * i / 150))
        offset_y = int(20.0 * np.cos(2 * 3.14 * j / 150))
        if i+offset_y < rows and j+offset_x < cols:
            img_both[i,j] = img[(i+offset_y)%rows,(j+offset_x)%cols]
        else:
            img_both[i,j] = 0

plt.subplot(224)
plt.imshow(img_both,cmap=plt.cm.gray)
plt.title('Both')
plt.axis('off')
        
#####################
# Concave effect

img_output = np.zeros(img.shape, dtype=img.dtype)

for i in range(rows):
    for j in range(cols):
        offset_x = int(128.0 * np.sin(2 * 3.14 * i / (2*cols)))
        offset_y = 0
        if j+offset_x < cols:
            img_output[i,j] = img[i,(j+offset_x)%cols]
        else:
            img_output[i,j] = 0
```

## 插值问题
   图像几何变换中采用逆变换函数主要是因为：数字图像的像素位置$(x,y)$是离散的，通过正变换后得到的像素位置$(x',y')$可能不是整数。采用拟变换可以根据变换后像素的位置$(x',y')$反向计算其在原图像的位置$(x,y)$，显然$(x,y)$也可能不是整数，此时可以采用插值的方式来获得非整像素位置的值。插值的方法主要有三种：
 * 最近邻
 * 双线性插值
 * 双三次插值
 下图表示插值像素的位置：
<img src= images/position.png width=50% height=50%>
   其中P表示需要插值像素的位置$(x,y)$，$Q_{11},Q_{12},Q_{21},Q_{22}$是整数点像素值。
   双线性插值涉及周围四个点
   双三次插值涉及周围16个点
   
   skimage的wrap函数有order参数，它定义了插值的方法的阶数，具体取值如下：

    0: Nearest-neighbor

    1: Bi-linear (default)

    2: Bi-quadratic

    3: Bi-cubic

    4: Bi-quartic

    5: Bi-quintic

## 非线性变换示例2
   这个例子展示了用skimage 实现wirl效果
### WIRL 效果原理
   $(x,y)$表示变换前的图像位置，$(x',y')$表示变换后的位置，$x_c,y_c$表示图像的中心位置。WIRL效果的变换公式为：
   $$
   \begin{align}
   d_x &= x'-x_c \\
   d_y &= y'-y_c \\
   r &= \sqrt{d_x^2+d_y^2} \\
   \beta &= arctan(d_y,d_x) + \alpha (\frac{r_{max}-r}{r_{max}})\\
   \end{align}
   $$
   
   $$
   \begin{align}
   x &= \left\{ \begin{array}{lr}x_c+rcos(\beta) & r<r_{max} \\
                                x' & r>r_{r_max}\\
               \end{array}
         \right. \\
   y &= \left\{ \begin{array}{lr}y_c+rsin(\beta) & r<r_{max} \\
                                y' & r>r_{r_max}\\
               \end{array}
         \right. \\
   \end{align}
   $$
   $r_{max},\alpha$是两个参数
   
   下面给出了skimage实现效果与PPT实现算法的对比

```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.transform import swirl


image = data.checkerboard()
swirled = swirl(image, rotation=23, strength=20, radius=120)

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3),
                               sharex=True, sharey=True)

ax0.imshow(image, cmap=plt.cm.gray)
ax0.axis('off')
ax1.imshow(swirled, cmap=plt.cm.gray)
ax1.axis('off')

plt.show()
```

```python
from skimage import io, data, color
import matplotlib.pylab as plt
import numpy as np
#inline matplotlib

img = data.checkerboard()
rows, cols = img.shape
yc = rows / 2   #图像中心位置
xc = cols / 2

rmax = 120
alpha = 4

io.imshow(img)
plt.subplot(121)
plt.imshow(img,cmap=plt.cm.gray)
plt.title('Original')
plt.axis('off')

#####################
# my wirl

img_wirl = np.zeros(img.shape, dtype=img.dtype)

for i in range(rows):
    for j in range(cols):
        dx = j - xc
        dy = i - yc
        r = np.sqrt(dx**2 +dy ** 2)
        beta = np.arctan(dy/(dx+1e-6)) + alpha * (rmax - r) / rmax
        if r <= rmax:
            x = xc + r * np.cos(beta)
            y = yc + r * np.sin(beta)
        else:
            x = j
            y = i
        if x < cols and y <rows:
            img_wirl[i,j] = img[int(y), int(x)]
        else:
            img_wirl[i,j] = 0
        
plt.subplot(122)
plt.imshow(img_wirl,cmap=plt.cm.gray)
plt.title('WIRL')
plt.axis('off')            
```

### Spherical transform 效果
   $(x,y)$表示变换前的图像位置，$(x',y')$表示变换后的位置，$x_c,y_c$表示图像的中心位置。Shperical transform效果的变换公式为：
   $$
   \begin{align}
   d_x &= x'-x_c \\
   d_y &= y'-y_c \\
   r &= \sqrt{d_x^2+d_y^2} \\
   z &= \sqrt{r_{max}^2-r^2} \\
   \beta_x &= (1-\frac{1}{\rho}) sin^{-1}(\frac{d_x}{\sqrt{r_{max}^2-d_y^2}})\\
   \beta_y &= (1-\frac{1}{\rho}) sin^{-1}(\frac{d_y}{\sqrt{r_{max}^2-d_x^2}})\\
   \end{align}
   $$
   
   $$
   \begin{align}
   x &= \left\{ \begin{array}{lr}x' - z\cdot tan(\beta_x) & for \ r<r_{max} \\
                                x' & r>r_{r_max}\\
               \end{array}
         \right. \\
   y &= \left\{ \begin{array}{lr}y' - z\cdot tany(\beta_y) & for \ r<r_{max} \\
                                y' & r>r_{r_max}\\
               \end{array}
         \right. \\
   \end{align}
   $$
   $r_{max},\rho$是两个参数，前者决定放大镜的半径，后者决定放大倍数。
   


```python
from skimage import io, data, color
import matplotlib.pylab as plt
import numpy as np
#inline matplotlib

img = data.checkerboard()
rows, cols = img.shape
yc = rows / 2   #放大镜中心位置
xc = cols / 2

rmax = 70
rho = 1.8

io.imshow(img)
plt.subplot(121)
plt.imshow(img,cmap=plt.cm.gray)
plt.title('Original')
plt.axis('off')

img_shperical = np.zeros(img.shape, dtype=img.dtype)

for i in range(rows):
    for j in range(cols):
        dx = j - xc
        dy = i - yc
        r = np.sqrt(dx**2 + dy ** 2)
        if r <= rmax: 
            z = np.sqrt(rmax ** 2 - r ** 2)
            beta_x = (1 - 1/rho) * np.arcsin(dx/(np.sqrt(dx ** 2 + z **  2) + np.finfo(np.float).eps))
            beta_y = (1 - 1/rho) * np.arcsin(dy/(np.sqrt(dy ** 2 + z **  2) + np.finfo(np.float).eps))
            x = j - z * np.tan(beta_x)
            y = i - z * np.tan(beta_y)
        else:
            x = j
            y = i
           
        img_shperical[i,j] = img[int(y),int(x)]
        
plt.subplot(122)
plt.imshow(img_shperical,cmap=plt.cm.gray)
plt.title('Sphere')
plt.axis('off')        
```
